#### Import Dependency

```python
from ultralytics import YOLO
import cv2 as cv
import cvzone
import os
import math
from sort import *
```


#### Loading Model


```python
model = YOLO("yolov8n-pose.pt")
```

```python
# Path to save clicked Photo of new individuals in frame
save_path = os.path.join(r"C:\Users\abdul\Downloads\PROJECTS\Deep_Learning_Project\Computer_vison\virt_env\Object_detection_project\preson_detected\Detected_photo")
```

```python
cap = cv.VideoCapture(0) # 0 for Camera
cap.set(3,1500) #set size of camera
cap.set(4,1000) #set size of camera
tracker = Sort(max_age=50, min_hits=3, iou_threshold=0.3) 
id_list = []
while True: 
    ret, img = cap.read() 
    results = model(img,stream=True)
    detections = np.empty((0,5))
    for result in results:
        boxe = result.boxes
        classes = result.names
        for box in boxe:
            x1,y1,x2,y2 = box.xyxy[0]
            x1,y1,x2,y2= int(x1),int(y1),int(x2),int(y2)
            w,h = x2-x1,y2-y1
            conf = box.conf[0]
            name = classes[int(box.cls[0])]
            if  conf > 0.55 :
                crr_arr = np.array([x1,y1,x2,y2,conf])
                detections= np.vstack([detections,crr_arr])
                
    result_track = tracker.update(detections)
    for track in result_track:
        x1,y1,x2,y2,ids = track
        x1, y1, x2, y2,ids = int(x1), int(y1), int(x2), int(y2),int(ids)
        w, h = x2 - x1, y2 - y1
        cx,cy =x1+w//2 ,y1+h//2
        cvzone.cornerRect(img,bbox=[x1,y1,w,h],t=2)
        cvzone.putTextRect(img,f"{ids} {name} {math.ceil(conf*100)} %",(max(0,x1),max(35,y1)),scale=1.3,thickness=1,
                               colorR=(120,81,169), font=1,)
        if ids not in id_list:
            id_list.append(ids)
            cv.imwrite(os.path.join(save_path,f"{ids}.jpg"),img)
            print("clicked")
    cv.imshow('Image', img)
    key = cv.waitKey(1)
    if key == ord('q'):  
        break
cap.release() # turn off camera
cv.destroyAllWindows() 
```
